```python
#doc -list, t_words,o_words - pd.dataFrame
#step - радиус окрашивания слов
def find_op(doc,t_words,o_words,to_file=False,vw_file=None,doc_id=None,step=5):

    t_doc=np.zeros(len(doc)) #массив длины докуманта из нулей.
    #если слово тематическое заменяем 0 на id слова  
    i=0
    tw1=Counter()
    tw2=Counter()
    for word in doc:
        t=t_words[t_words[0]==word]
        
        if len(t.index)!=0: #если слово тематичекое, то добавить его индекс
            t_doc[i]=t.index[0]
            if t[1].values[0]=="text_ngramm":
                tw1=tw1+Counter([word])
            else:
                tw2=tw2+Counter([word])
        i+=1
    
    #записываем в файл
    if to_file==True:
        doc_info='doc'+`doc_id`+" |text_ngramm "     
        for w in tw1:
            doc_info=doc_info+str(w)+":"+str(tw1[w])+" "
        doc_info=doc_info+" |title_ngramm "
        for w in tw2:
            doc_info=doc_info+str(w)+":"+str(tw2[w])+" "
        
    
    #поиск тональных слов в документе
    pol_doc=np.zeros(len(doc))
    i=0

    for word in doc:
        f=o_words[o_words['0']==word.decode('utf-8')]
        if len(f)!=0:
            pol_doc[i]=f['1'].values[0]
        i+=1
        
    #распределеие тональности на тематические слова в радиусе 5
    posw=Counter()
    negw=Counter()
    op_dict={}
    for i in range(len(pol_doc)):
        if pol_doc[i]!=0:
            l=i-step
            if l<0:
                l=0
            for j in range(len(t_doc[l:i+step])):
                if t_doc[l+j]!=0:
                    if (pol_doc[i]==1):
                        posw+=Counter([t_words.ix[[t_doc[l+j]]][0].values[0]])
                    else:
                        negw+=Counter([t_words.ix[[t_doc[l+j]]][0].values[0]])

                    if (t_words.ix[[t_doc[l+j]]][0].values[0] not in op_dict ):
                        op_dict[t_words.ix[[t_doc[l+j]]][0].values[0]]=pol_doc[i]
                    else:
                        op_dict[t_words.ix[[t_doc[l+j]]][0].values[0]]+=pol_doc[i]
                        
    if to_file==True:
        doc_info=doc_info+"|neg_pol "
        for w in negw:
            doc_info=str(doc_info)+" "+str(w)+":"+str(negw[w])
            
        doc_info=doc_info+" |pos_pol "
        for w in posw:
            doc_info=str(doc_info)+" "+str(w)+":"+str(posw[w])
        doc_info+="\n"
        vw_file.write(doc_info)
    return op_dict
```

#### Модель 

```python
T = 5   # количество тем
model = artm.ARTM(num_topics=T, 
                  topic_names=["t"+str(i) for i in range(T)], 
                  dictionary=dictionary,
                  class_ids={'title_ngramm':3,'text_ngramm':5,'pos_pol':5,'neg_pol':5}, 
                  seed=-1)
```

```python
model.scores.add(artm.PerplexityScore(name='PerplexityScore',
                                               use_unigram_document_model=False,
                                               dictionary=dictionary))
model.scores.add(artm.SparsityPhiScore(name='SparsityPhiScore', class_id="text"))
model.scores.add(artm.SparsityThetaScore(name='SparsityThetaScore'))
model.scores.add(artm.TopTokensScore(name="top_words", num_tokens=30, class_id="text_ngramm"))
model.scores.add(artm.TopTokensScore(name="top_pos_words", num_tokens=30, class_id='pos_pol'))
model.scores.add(artm.TopTokensScore(name="top_neg_words", num_tokens=30, class_id='neg_pol'))
```

```python
model.regularizers.add(artm.SmoothSparsePhiRegularizer(name='SparsePhi', tau=-9*1e4, dictionary="dictionary",class_ids=['text_ngramm']))
model.regularizers.add(artm.DecorrelatorPhiRegularizer(name='DecorrelatorPhi', tau=800,class_ids=['text_ngramm']))

model.regularizers.add(artm.SmoothSparsePhiRegularizer(name='pos_neg_sparsePhi', tau=-9*1e4, dictionary="dictionary",class_ids=['pos_pol','neg_pol']))
model.regularizers.add(artm.DecorrelatorPhiRegularizer(name='pos_neg_decorrelatorPhi', tau=800,class_ids=['pos_pol','neg_pol']))
```

```python
model.fit_offline(batch_vectorizer=batch_vectorizer, num_collection_passes=25)
```

```python
last= model.score_tracker['top_pos_words'].last_tokens
key=last.keys()

for j in key:
    print j+': ',
    for k in last[j]:
        print k,        
    print
```
